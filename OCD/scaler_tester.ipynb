```python
import math

def easeOutBack(t, b=1, c=7):
    return c * (1 - (math.sin(t * (2 - t) * math.pi / 2)) ** 2.7) + b

def calculate_replicas(rps):
    # Normalize rps into the range [0, 1]
    normalized_rps = min(1, rps / (8 * 15))

    # Calculate the replicas using the easeOutBack easing function
    replicas = easeOutBack(normalized_rps)

    # Ensure the replicas are in the range [1, 8]
    replicas = max(1, min(8, replicas))

    return int(replicas)

# Example usage:
rps = 60
print("Replicas:", calculate_replicas(rps))  # Output will be 3
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math
import pandas as pd

def easeOutBack(t, b=1, c=7):
    return c * ((math.sin(t * (2 - t) * math.pi / 2)) ** 1) + b

def calculate_replicas(rps):
    normalized_rps = min(1, rps / (8 * 12))
    replicas = easeOutBack(normalized_rps)
    replicas = max(1, min(8, replicas))
    return int(replicas)

# Create a DataFrame to hold the rps and replicas values
data = {'rps': [], 'replicas': []}
for rps in range(0, 100):
    data['rps'].append(rps)
    data['replicas'].append(calculate_replicas(rps))

df = pd.DataFrame(data)

# Plot the data using Seaborn
sns.lineplot(x='rps', y='replicas', data=df)
plt.title('Replicas vs RPS')
plt.xlabel('Requests Per Second (RPS)')
plt.ylabel('Number of Replicas')
plt.show()
```
